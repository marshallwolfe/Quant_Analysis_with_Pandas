## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")

# Create dataframe
whale_returns_df = pd.read_csv(whale_returns_csv, index_col = "Date", infer_datetime_format=True, parse_dates=True)

# Display a few rows
whale_returns_df.head()
```

```python
# Count nulls

# Checking for null
whale_returns_df.isnull()
# Determining number of nulls
whale_returns_df.isnull().sum()
```

```python
# Drop nulls

# Use the `dropna` function to drop whole records that have at least one null value
whale_returns_df.dropna(inplace=True)
whale_returns_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")

# Create dataframe
algo_returns_df = pd.read_csv(algo_returns_csv, index_col = "Date", infer_datetime_format=True, parse_dates=True)

# Display a few rows
algo_returns_df.head()
```

```python
# Count nulls

# Checking for null
algo_returns_df.isnull()
# Determining number of nulls
algo_returns_df.isnull().sum()
```

```python
# Drop nulls

# Use the `dropna` function to drop whole records that have at least one null value
algo_returns_df.dropna(inplace=True)
algo_returns_df.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")

# Create and sort dataframe
sp500_history_df = pd.read_csv(sp500_history_csv, index_col = "Date", infer_datetime_format=True, parse_dates=True)
sp500_history_df.sort_values(by=['Date'], inplace=True)

# Display a few rows
sp500_history_df.head()
```

```python
# Check Data Types

# Use the `dtypes` attribute to output the column names and data types
sp500_history_df.dtypes
```

```python
# Fix Data Types

# Use the `str.replace` method to remove dollar signs
sp500_history_df['Close'] = sp500_history_df['Close'].str.replace('$','')

# Use the `as_type` function to convert `Close` from `object` to `float`
sp500_history_df['Close'] = sp500_history_df['Close'].astype('float')

# Use the `dtypes` attribute to output the column names and data types
sp500_history_df.dtypes
```

```python
# Calculate Daily Returns

# Calculate the daily return using the 'pct_change()' function
sp500_daily_returns_df = sp500_history_df.pct_change()
sp500_daily_returns_df.head()
```

```python
# Drop nulls

# Use the `dropna` function to drop whole records that have at least one null value
sp500_daily_returns_df.dropna(inplace=True)
sp500_daily_returns_df.head()
```

```python
# Rename Column

sp500_daily_returns_df = sp500_daily_returns_df.rename(columns={"Close": "S&P 500"})
sp500_daily_returns_df.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame

# Concatenate the DataFrames by columns and perform an inner join
combined_df = pd.concat([whale_returns_df, algo_returns_df, sp500_daily_returns_df], axis="columns", join="inner")
combined_df.head()
```

## Performance

Calculate and Plot the daily returns and cumulative returns. Does any portfolio outperform the S&P 500?

Determine the _risk_ of each portfolio:

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
2. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window

combined_df.rolling(window=21).std().plot(figsize=(20,10), title="21 Day Rolling Standard Deviation")
```

### It does appear from the 21 Day Rolling Standard Deviation chart that risk increases for each of the portfolios at the same time as risk increases for the S&P 500.

```python
# Construct a correlation table

combined_df.corr()
```

### Algo 2 appears to be most closely correlated with the S&P 500 with a correlation coefficient of 0.858764 (the largest of the group), closely followed by Soros Fund Management LLC with a correlation coefficient of 0.837864. 

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
# (Your graph may differ, dependent upon which portfolio you are comparing)

# Calculate covariance of all daily returns of Algo 1 vs. S&P 500
covariance = combined_df['Algo 1'].cov(combined_df['S&P 500'])
# covariance
# Calculate variance of all daily returns of S&P 500
variance = combined_df['S&P 500'].var()
# variance
# Calculate beta of all daily returns of Algo 1
algo1_beta = covariance / variance
print(f"The Beta for the Algo 1 portfolio is {algo1_beta}.")
```

```python
# Calculate 21-day rolling covariance of Algo 1 vs. S&P 500 and plot the data
rolling_covariance = combined_df['Algo 1'].rolling(window=21).cov(combined_df['S&P 500'])
rolling_covariance.plot(figsize=(20, 10), title='Rolling 21-Day Covariance of Algo 1 Returns vs. S&P 500 Returns')
```

```python
# Calculate 21-day rolling variance of S&P 500 and plot the data
rolling_variance = combined_df['S&P 500'].rolling(window=21).var()
rolling_variance.plot(figsize=(20, 10), title='Rolling 21-Day Variance of S&P 500 Returns')
```

```python
# Calculate 21-day rolling beta of Algo 1 and plot the data
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 21-Day Beta of Algo 1')
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

Calculate and plot the annualized Sharpe ratios for all portfolios to determine which portfolio has the best performance

```python
# Annualized Sharpe Ratios

sharpe_ratios = (combined_df.mean() * 252) / (combined_df.std() * np.sqrt(252))
sharpe_ratios
```

 plot() these sharpe ratios using a barplot.
 On the basis of this performance metric, do our algo strategies outperform both 'the market' and the whales?

```python
# Visualize the sharpe ratios as a bar plot

sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

### Algo 1 has the highest sharpe ratio so it has the highest return per unit of risk of all the analyzed portfolios. The S&P 500 has the next highest sharpe ratio, followed by Berkshire Hathaway, followed by Algo 2. Soros Fund Management LLC, Paulson & Co Inc, and Tiger Global Management LLC trailed Algo 2. Tiger Global Management LLC and Paulson & Co Inc have negative sharpe ratios.

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock
goog_historical_csv = Path("Resources/goog_historical.csv")

# Create a dataframe
goog_historical_df = pd.read_csv(goog_historical_csv, index_col = "Trade DATE", infer_datetime_format=True, parse_dates=True)

# Sort the dataframe
goog_historical_df.sort_values(by=['Trade DATE'], inplace=True)

# Display a few rows
goog_historical_df.head()
```

```python
# Read the second stock
aapl_historical_csv = Path("Resources/aapl_historical.csv")

# Create a dataframe
aapl_historical_df = pd.read_csv(aapl_historical_csv, index_col = "Trade DATE", infer_datetime_format=True, parse_dates=True)

# Sort the dataframe
aapl_historical_df.sort_values(by=['Trade DATE'], inplace=True)

# Display a few rows
aapl_historical_df.head()
```

```python
# Read the third stock
cost_historical_csv = Path("Resources/cost_historical.csv")

# Create a dataframe
cost_historical_df = pd.read_csv(cost_historical_csv, index_col = "Trade DATE", infer_datetime_format=True, parse_dates=True)

# Sort the dataframe
cost_historical_df.sort_values(by=['Trade DATE'], inplace=True)

# Display a few rows
cost_historical_df.head()
```

```python
# Concatenate all stocks into a single DataFrame

custom_df = pd.concat([goog_historical_df, aapl_historical_df, cost_historical_df], axis="rows", join="inner")
custom_df.head()
```

```python
# Reset the index

custom_df = custom_df.reset_index()
custom_df.head()
```

```python
# Pivot so that each column of prices represents a unique symbol

pivot_df = custom_df.pivot(index='Trade DATE', columns='Symbol', values='NOCP')
pivot_df.head()
```

```python
# Drop Nulls

pivot_df = pivot_df.pct_change()
pivot_df.dropna(inplace=True)
pivot_df.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]

# Use the dot method to calculate weighted portfolio returns
pivot_df.dot(weights).head()
```

## Join your custom portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns

all_returns_df = pd.concat([combined_df, pivot_df.dot(weights)], axis="columns", join="inner")
all_returns_df.columns = ['SOROS FUND MANAGEMENT LLC', 'PAULSON & CO.INC.', 'TIGER GLOBAL MANAGEMENT LLC', 'BERKSHIRE HATHAWAY INC', 'Algo 1', 'Algo 2', 'S&P 500', 'Custom']
all_returns_df.tail()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)

all_returns_df.isnull().sum()
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others

```python
# Risk

new_daily_std = all_returns_df.std()
new_annualized_std = new_daily_std * np.sqrt(252)
new_annualized_std
```

```python
# Rolling

all_returns_df.rolling(window=21).std().plot(figsize=(20,10), title="21 Day Rolling Standard Deviation")
```

```python
# Beta

# Calculate covariance of all daily returns of Custom vs. S&P 500
new_covariance = all_returns_df['Custom'].cov(all_returns_df['S&P 500'])

# Calculate variance of all daily returns of S&P 500
new_variance = all_returns_df['S&P 500'].var()

# Calculate beta of all daily returns of Custom
custom_beta = new_covariance / new_variance
print(f"The Beta for the Custom portfolio is {custom_beta}.")
```

```python
# Calculate 21-day rolling covariance of Custom vs. S&P 500 and plot the data
new_rolling_covariance = all_returns_df['Custom'].rolling(window=21).cov(all_returns_df['S&P 500'])
new_rolling_covariance.plot(figsize=(20, 10), title='Rolling 21-Day Covariance of Custom Returns vs. S&P 500 Returns')
```

```python
# Calculate 21-day rolling variance of S&P 500 and plot the data
new_rolling_variance = all_returns_df['S&P 500'].rolling(window=21).var()
new_rolling_variance.plot(figsize=(20, 10), title='Rolling 21-Day Variance of S&P 500 Returns')
```

```python
# Calculate 21-day rolling beta of Custom and plot the data
new_rolling_beta = new_rolling_covariance / new_rolling_variance
new_rolling_beta.plot(figsize=(20, 10), title='Rolling 21-Day Beta of Custom')
```

```python
# Annualized Sharpe Ratios

new_sharpe_ratios = (all_returns_df.mean() * 252) / (all_returns_df.std() * np.sqrt(252))
new_sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot

new_sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

### Algo 1 has the highest sharpe ratio so it has the highest return per unit of risk of all the analyzed portfolios. The Custom portfolio has the next highest sharpe ratio, followed by the S&P 500. Soros Fund Management LLC, Paulson & Co Inc, and Berkshire Hathaway trail the S&P 500. Tiger Global Management LLC has the lowest sharpe ratio which is negative.
